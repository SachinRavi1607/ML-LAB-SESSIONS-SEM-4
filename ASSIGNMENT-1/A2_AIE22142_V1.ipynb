```python
import math

def dist_euclidean(v1, v2):
    if len(v1) != len(v2):
        raise ValueError("Vector dimensions must match")

    dist = 0
    for i in range(len(v1)):
        dist += (v1[i] - v2[i])**2
    return math.sqrt(dist)


def classify_k_nearest(train_data, test_inst, k):
    distances = []
    for train_inst, label in train_data:
        distance = dist_euclidean(train_inst, test_inst)
        distances.append((distance, label))

    distances.sort(key=lambda x: x[0])

    neighbors = distances[:k]

    class_votes = {}
    for _, label in neighbors:
        class_votes[label] = class_votes.get(label, 0) + 1

    return max(class_votes, key=class_votes.get)


if __name__ == "__main__":
   
    v1 = [1, 2, 3]
    v2 = [4, 5, 6]
    print("Euclidean distance:",dist_euclidean(v1, v2))

    data = [([1, 2], 'A'), ([2, 3], 'B'), ([3, 4], 'A')]
    test_inst = [1.5, 2.5]
    k = 2
    print("Predicted class label:", classify_k_nearest(data, test_inst, k))

```

```python
import random

def label_encoding(inputdata):
    encoding_map_dictionary = {}
    encoded_data = []

    random.seed(50)

    for obj in inputdata:
        if obj not in encoding_map_dictionary:
            encoding_map_dictionary[obj] = int(random.random() * 1000)  
        encoded_data.append(encoding_map_dictionary[obj])
    
    return encoded_data

def read_dataset(filename):
    dataset = []
    with open(filename, 'r') as file:
        lines = file.readlines()
        for line in lines[1:]:
            row = line.strip().split(',')
            dataset.append(row)
    return dataset

def main():
    filename = "C:\\Users\\SACHIN.R\\Downloads\\archive\\cars_ds_final_2021.csv"
    cars_dataset = read_dataset(filename)

    categorical_column = [row[15] for row in cars_dataset]
    encoded_column = label_encoding(categorical_column)

    print("Original species data:", categorical_column)
    print("Encoded species data:", encoded_column)

if __name__ == "__main__":
    main()
```

```python
def read_csv(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        header = lines[0].strip().split(',')
        data = [line.strip().split(',') for line in lines[1:]]
    return header, data

def print_dataframe(header, data):
    print(','.join(header))
    for row in data:
        print(','.join(row))

def one_hot_encode(column):
    unique_values = list(set(column))
    encoding_dict = {value: i for i, value in enumerate(unique_values)}
    
    encoded_data = []
    
    for value in column:
        encoding = [0] * len(unique_values)
        encoding[encoding_dict[value]] = 1
        encoded_data.append(encoding)
    
    return encoded_data

# Example usage:
file_path = 'C:\\Users\\SACHIN.R\\Downloads\\archive\\cars_ds_final_2021.csv'
header, data = read_csv(file_path)

# Choose a categorical column for one-hot encoding
categorical_column_name = 'Drivetrain'
column_index = header.index(categorical_column_name)
categorical_column = [row[column_index] for row in data]

# Apply one-hot encoding
encoded_data = one_hot_encode(categorical_column)

# Add the one-hot encoded columns to the data
for i, unique_value in enumerate(set(categorical_column)):
    new_column_name = f"{categorical_column_name}_{i}"
    header.append(new_column_name)
    new_column = [row[i] for row in encoded_data]
    for j in range(len(data)):
        data[j].append(str(new_column[j]))

# Print the updated dataset
print_dataframe(header, data)
```
